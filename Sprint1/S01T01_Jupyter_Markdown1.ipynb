### By using Montecarlo integration get the value of Pi.

We will infer the Pi value by using the fact we have a square of $Area=2*2$ with inscribed circle of radius 1, which $A_{circle} = \pi r^2 \xrightarrow{r=1} \pi$. 

If we throw two random values in a square of $[ -1,1 ]$ and calculate the distance: $d=x^2 + y^2$ the value will belong in the circle if it's less than 1. Since we pick the range  $[ -1,1 ]$, the complete formula to obtain pi is: $$\pi = 4*\frac{\mbox{num values in the circle}}{\mbox{num total values generated}}$$


```python
import numpy as np
import math

pi_real = math.pi
# distance
d = 0
# computed value of pi
pi = 0
# number of iteracions
n = 1
# how small the difference between the real number and computed is
epsilon = 1e-6
while abs(pi_real - pi) > epsilon: 
    x = np.random.uniform(-1,1)
    y = np.random.uniform(-1,1)
    
    if (x**2 + y**2) < 1:
        d += 1

    n += 1
    pi = 4*d / n

print('Computed Pi value = ',pi)
print('Number of iteraccions = ', n)
```

### By using Euler integration get the Lorentz attractors

The program below computes the dynamic equations for Lorentz attractors by using the integration method of Euler.

They are called dynamic equations Lorentz, since it was Lorentz who wrote $a=10$, $b=28$, and $c=8/3$ (<a href="https://mathworld.wolfram.com/LorenzAttractor.html">Wolfram Alpha</a>)

I have suppressed the result, while it shows the graph resultant of the computed values.

```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

# constant values
a = 10
b = 28
c = 8/3

# time step we want to increase each variable, and number of iterations.
dt = 0.01
n = 10000

#initialize values:
x = [0 for i in range(n+1)]
y = [0 for i in range(n+1)]
z = [0 for i in range(n+1)]

x[0], y[0], z[0] = (0.1, 0, 0)

# Use Euler Integration to determine the new values:
i = 0
while i < n:
    # Get the Lorentz the derivaties
    xdot = a*(y[i] - x[i])
    ydot = x[i]*(b - z[i]) - y[i]
    zdot = x[i]*y[i] - c*z[i]

   # update value
    x[i+1] = x[i] + xdot*dt
    y[i+1] = y[i] + ydot*dt
    z[i+1] = z[i] + zdot*dt
    
    i += 1

#print(x,y,z)

# plot
fig = plt.figure()
ax = plt.axes(projection ='3d')

ax.plot3D(x,y,z)
ax.set_title("Lorenz Attractor")
plt.show()
```
