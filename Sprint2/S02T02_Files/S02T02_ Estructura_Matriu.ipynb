## Exercici 4: Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

In the documentation says that broadcasting is how python does arithmetic calculations with arrays 
with different shapes. It keeps the dimension of the highest array, it is also known as "element-wise operation" (ex. if we have a 3d array against 1d the result will be a 3d array).

In the case of the multiplication, in programming this is called multiplication, while the mathematical multiplication of two elements of different dimensions (understanding that they are dimensionally compatible to do operations $(n \times m)*(m \times l)$) in programming is called inner product.

https://numpy.org/doc/stable/user/basics.broadcasting.html.

(In this examples I will using integers, arrays of floats behave the same.)

```python
import numpy as np
# taking the basic example of an array multiplied or sumed by an scalar:
a = np.array([1,2,3,4]) # array 1d
b = 3 # scalar
```

```python
# sum: adds the scalar b = 3 to each one of the values on the array. In this case: [4,5,6,7]
a + b 
```

```python
# multiplication: multiplies the scalar b for each one of values on the array. In this case: [3,6,9,12]
print(a*b)
```

```python
#Now we take two 1d arrays, in which, of course, the dimensions are the same.
a = np.array([1,2,3])
b = np.array([1,2,3])
```

```python
# sum: the direct sum the two arrays: [2,4,6]
a+b
```

```python
#multiplication: multiplies each element of the vector by element (element-wise) in the same position 
#of the other vector, resulting: [1, 4, 9]
a*b
```

The mathematical multiplication of two vectors (a and b) that results into a scalar, called inner product, is given in python by the function np.dot(a,b). In the case of the last example we will have that the result is instead 14.

```python
np.dot(a,b)
```

```python
# Additionaly we can run the same with 2d arrays. The same result applies for bigger dimensions. 
# In https://numpy.org/doc/stable/user/basics.broadcasting.html, on the section of Broadcasting arrays we find
# more examples for distinct dimensions.

#1. matrix against scalar
a = np.array([[1,2],[1,2]])
b = 3 # scalar

print('suma array-escalar:') 
print(a+b)
print('multiplicacio array-escalar:')
print(a*b)

#2. matrix against matrix
a = np.array([[1,2,3],[1,2,3]])
b = np.array([[1,2,3],[1,2,3]])

print('suma array-array:') 
print(a+b)
print('multiplicacio array-array:')
print(a*b)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image = mpimg.imread('IMG_20191213_200336.jpg')
plt.imshow(image)
```

```python
# find the image dimensions aka the matrix dimensions:
print('Dimensions de la imatge:',image.shape)
```

Podem veure que la image es una matriu 3dimensional a on la tercera dimensio, que te dimensio 3, cadascuna de les matriu correspon als color vermell, verd i blau. Aixi doncs, la indexacio image[:,:,i], a on i = 0,1,2 per vermell, verd, i, blau correspondent. 

```python
# we can print each one of them
print('Matriu que dona els vermells a la imatge:')
print(image[:,:,0])

print('Matriu que dona els verds a la imatge:')
print(image[:,:,1])

print('Matriu que dona els blaus a la imatge:')
print(image[:,:,2])
```

Com podem veure aquestes matrius tenen valors que van del 0 al 255, que corresponen al "arc" RGB. Modificant algun d'aquests valors canvia el color de la imatge.

```python
# L'exercici ens demana modificar algun dels valors de les matrius i mirar que hi passa a la imatge.
# En aquest cas, mirem que passa quan posem un filtre a la matriu del vermells, per exemple:
new_image = (image[:,:,0] > 150) #nomes agafem els valors del vermell superiors a 150!
plt.imshow(new_image) # la imatge ens mostra verds i blaus
```

```python
# guardem la imatge:
mpimg.imsave('filtered', new_image, format='jpg')
```

Amb imshow, una de les moltes opcions, es poder representar la imatge en cadascun dels colors RGB. Altres opcions ens permeten fins i tot canviar els colors a mes calids (cmap='hot') o veure la frequencia per canviar-ne el contrast (referencies [1] i [2]).
