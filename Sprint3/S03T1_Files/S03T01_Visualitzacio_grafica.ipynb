## Exercici 1

```python
# load dataset
df1 = pd.read_csv('DelayedFlights.csv')
```

### Resumeix gràficament el data set DelayedFlights.csv. Crea almenys una visualització per:

### 1. Una variable categòrica (UniqueCarrier)

Per fer el plot de una sola variable triem la funcio hist() de la llibreria matplotlib.

```python
plt.ylabel('frequency')

plt.hist(df1['UniqueCarrier'])
plt.savefig('graph31.png')

plt.show()
```

### 2. Una variable numèrica (ArrDelay)

Aqui fem el mateix que abans pero pel cas en que la variable es numerica.

```python
plt.ylabel('frequency')
plt.xlabel('minutes')

plt.hist(df1['ArrDelay'])
plt.savefig('graph32.png')
plt.show()
```

### 3.Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)

Ara utilitzarem la lliberia seaborn.

```python
sns.scatterplot(data=df1, x='UniqueCarrier', y='ArrDelay')

plt.savefig('graph33.png')
```

### 4. Dues variables numèriques (ArrDelay i DepDelay)

```python
sns.scatterplot( data=df1, x='ArrDelay', y='DepDelay')

plt.savefig('graph34.png')
```

### 5.Tres variables (ArrDelay, DepDelay i UniqueCarrier)

```python
sns.scatterplot(data=df1, x='ArrDelay', y='DepDelay', hue='UniqueCarrier')
# set legend outside the plot, since it takes to space.
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)

plt.savefig('graph35.png')
```

### 6. Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier).

```python
# seleccionem el subset de variables que volem representar.
df1b = df1[['ArrDelay', 'DepDelay', 'AirTime', 'UniqueCarrier']]
sns.pairplot(df1b, hue ='UniqueCarrier')

plt.savefig('graph36.png')
```
