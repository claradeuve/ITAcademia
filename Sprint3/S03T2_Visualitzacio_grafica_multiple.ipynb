### <span style=color:blue>STATISTICS</span>

Podem descriure matematicament, sense saber la grafica de una variable aleatoria amb un set en un determinat interval, a partir dels seus cumulants. Els quatre primers es refereixen a la mitjana, variancia, asimetria i curtosis. Aquestes mesures en donen el valor del centre del data set, la seva discrepancia (arrel quadrada de la varianca es la desviacio estandar), com de asimetrica es la grafica del set (skewness), i com es la cua d'aquesta grafica (curtosis).

Aixo es pot fer ja que els cumulants defineixen de manera compacte la funcio de densitat de probabilitat (PDF), que es la funcio que segueix una variable random $x \in \mathbb{R}$ en l'interval: $\mathcal{S}_{x} = \{ -\infty < x < \infty \}$.

$$prob(\mathcal{S}) = \int_{-\infty}^{\infty} dx p(x)$$

A partir de la funcio characteristica de la PDF, que es la transformada de Fourier de la PDF,

$$\tilde{p}(x) \equiv < e^{-ikx} > = \int dx p(x) e^{-ikx}$$

la qual treballant-la arribem a l'expressio de la funcio generatriu per els cumulants de la PDF:

$$e^{ikx_0}\tilde{p}(x) = \sum_{n=0}^{\infty} \frac{(-ik)^n}{n!}<(x-x_0)>\Rightarrow ln\tilde{p}(x) = \sum_{n=1}^{\infty}\frac{(-ik)^n)}{n!}<x>_{c}$$

Substituint $n=1,2,3,4$ obtenim els quatre primers cumulants els quals son referits com a Mitjana, Varianca, Asimetria i curtosis.

[1] M. Kardar - Statistical Physics of Particles. Cambridge University Press (2007)

Load data:

```python
movies = pd.read_csv('movies.dat', sep='::', header=None, 
                     names=['Id','title(year)','genre'], 
                    encoding='iso-8859-16')
```

```python
movies.head()
```

Now we need to split the name of the movie from the year. To do so, we need a regular expression that takes into account the paranthesis.

To extract the year we will use a regular expression that says that were are parenthesis and numerical values we extract the numericals. While to extract the movie title we can be a bit dificult, since we need a regular expression that takes into account, more than one word, apostrofs, diaresis, etc. To simplify, we can see that always, in the dataset, there are 2 parantheis + 4 numbers + 1 space = 7 characters, thus we sustract them.


```python
movies['Year'] = movies['title(year)'].str.extract('.*\((\d+)\).*')
movies['Title'] = movies['title(year)'].str[:-7]


movies.head()
```

```python
movies.drop('title(year)', axis=1, inplace=True)
```

We can see in the graph that the huge increment of movies from the 90s to the 2k is cause by the high production of movies in the genres: Drama and Children, mostly. The increase of films on the genres of Romance, Action and thriller, helped to have this peak of releases during the last half of the 90s.

```python
# Decade column:
movdf['Decade'] = movdf.Year.sub(min(movdf.Year)+1)//10

# we need to take into account the movies from the 1919. One way is making a new decade: 1910s, but since we only
# have one year, we will add the 1919 to the 20s decade.
#movdf['Decade'] = movdf.Decade + 1
movdf.Decade.loc[movdf.Year == min(movdf.Year)] = 0

#we do the same for the movies released on 2000. Instead, we add them on the decade of 90s
movdf.Decade.loc[movdf.Year == max(movdf.Year)] = 7
```

```python
# Decades:
sns.set(style='white',rc={'figure.figsize':(11,8)})
sns.scatterplot(data = movdf, x='Decade', y='Year')
plt.show()
```

```python
#find range of decades
n = movdf.Decade.nunique()
for i in range(0,n):
    mi = min(movdf.Year.loc[movdf.Decade == i]) 
    ma = max(movdf.Year.loc[movdf.Decade == i])
    print('Decade {}: {}-{}'.format(i, mi, ma))
    

movdf['Decade'] = movdf['Decade'].apply(lambda x: '{}0s'.format(x+2))    
```

So, we have split the data in 8 "decades", since we want to take into account the movies from the 1919 to the 20s and the ones released on 2000 to the 90s.

```python
# mean releases by genre and by decada.
mmov = movdf.groupby('Decade')[['Action', 'Adventure', 'Animation', "Children's",
       'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir',
       'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War',
       'Western']].mean()
mmov.reset_index(inplace=True)
mmov
```

```python
# standard deviation releases by genre and by decada.
smov = movdf.groupby('Decade')[['Action', 'Adventure', 'Animation', "Children's",
       'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir',
       'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War',
       'Western']].std()
smov.reset_index(inplace=True)
smov
```

```python
# plot mean values and std values as error bars.
dfm = mmov.melt(id_vars=['Decade'], value_vars=['Action', 'Adventure', 'Animation', "Children's",
       'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir',
       'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War',
       'Western'], var_name='genre', value_name='mean')
# The code down here does the same for dfs and then concatenates both subdatasets.
dfs = smov.melt(id_vars=['Decade'], value_vars=['Action', 'Adventure', 'Animation', "Children's",
       'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir',
       'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War',
       'Western'], var_name='genre', value_name='std')

dfms = pd.concat([dfm, dfs['std']], axis=1, join='inner')
```

```python
sns.set(style='white',rc={'figure.figsize':(11,8)})
sns.pointplot(x='genre', y='mean', hue='Decade', data=dfms)
plt.xticks(rotation=90)
plt.show()
```

In all decades the most populars films to shot are the Comedies and Dramas. While the least popular are Crime, Documentary, Fantasy, Film-Noir.

```python
sudec = movdf.groupby('Decade')[[ 'Action', 'Adventure', 'Animation', "Children's",
       'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir',
       'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War',
       'Western']].sum()
sudec.reset_index(inplace=True)
sudec.drop('Decade', axis=1)
```

```python
# find the probability of each genre:
probdf = sudec
cols = [ 'Action', 'Adventure', 'Animation', "Children's",
       'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir',
       'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War',
       'Western']
probdf[cols] = probdf[cols].div(probdf[cols].sum(axis=1), axis=0)

probdf
```

```python
dfp = probdf.melt(id_vars=['Decade'], value_vars=['Action', 'Adventure', 'Animation', "Children's",
       'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir',
       'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War',
       'Western'], var_name='genre', value_name='prob')
```

```python
sns.catplot(x='Decade', y='prob', col='genre', col_wrap=4, data=dfp, kind='point')
plt.show()
```

```python
sns.catplot(x='Decade', y='prob', hue='genre', data=dfp, kind='point')
plt.show()
```

We can see the probabilities of relasing movies by the genre. We also notice that while in some decades some generes are popular in other decades are not so much.
